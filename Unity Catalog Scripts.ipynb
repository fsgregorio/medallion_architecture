```sql
-- Create a ecommerce Catalog
CREATE CATALOG IF NOT EXISTS ecommerce_data
COMMENT 'E-commerce data catalog for Medallion Architecture practice';
```

```sql
-- Bronze Layer
CREATE SCHEMA IF NOT EXISTS ecommerce_data.bronze
COMMENT 'Bronze layer: Raw data with quality issues';

-- Silver Layer
CREATE SCHEMA IF NOT EXISTS ecommerce_data.silver
COMMENT 'Silver layer: Cleaned and validated data';

-- Gold Layer
CREATE SCHEMA IF NOT EXISTS ecommerce_data.gold
COMMENT 'Gold layer: Business-level aggregations and star schema';
```
